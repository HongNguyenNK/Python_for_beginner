```python
import pandas as pd
import matplotlib.pyplot as plt
#import numpy as np 

df = pd.read_csv('E:/Documences/py/Dr_Tram/ensembles-calib-raw.txt',delimiter='\t')
m, n = df.shape

fig=plt.plot(df['Year'],df.iloc[:,1:],'-.', lw = 0.8) 
df['mean']=df.iloc[:,1:].mean(axis=1)
#fig=plt.plot(df['Year'],df.iloc[:,1:-2],'.',color='grey',alpha=0.1)
fig=plt.plot(df['Year'],df['mean'],'k', linestyle ='-',lw=1.2, alpha = 1, label = 'Mean (ensembles-calib-raw)')   
```

```python
df10 = pd.read_csv('E:/Documences/py/Dr_Tram/ensembles-10yearsmth.txt',delimiter='\t')
m10, n10 = df10.shape

#fig=plt.plot(df10['Year'],df10.iloc[:,1:],'-.', lw = 0.8) 
df10['mean10']=df10.iloc[:,1:].mean(axis=1)
#fig=plt.plot(df['Year'],df.iloc[:,1:-2],'.',color='grey',alpha=0.1)
fig=plt.plot(df10['Year'],df10['mean10'],'blue', linestyle ='-',lw=1.0, alpha = 1, label = 'Mean (ensembles-10yearsmth)')
```

```python
df50 = pd.read_csv('E:/Documences/py/Dr_Tram/ensembles-50yearsmth.txt',delimiter='\t')
m50, n50 = df50.shape

#fig=plt.plot(df10['Year'],df10.iloc[:,1:],'-.', lw = 0.8) 
df50['mean50']=df50.iloc[:,1:].mean(axis=1)
#fig=plt.plot(df['Year'],df.iloc[:,1:-2],'.',color='grey',alpha=0.1)
fig=plt.plot(df50['Year'],df50['mean50'],'lime', linestyle ='-',lw=1.0, alpha = 1, label = 'Mean (ensembles-50yearsmth)')




plt.axhline(y=0, color='r', linestyle='--', lw= 0.9)   
plt.grid(linestyle ='-.')
plt.xlabel('Year')
plt.ylabel('Temp. Anomalies ($^o$C)')
plt.legend() 
plt.title('Ensembles calibration raw data by Frank (2010)')
plt.show() 
```
