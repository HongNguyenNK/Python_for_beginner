```python
import matplotlib.pyplot as plt
import pandas as pd


file = "E:/Documences/py/Dr_Tram/Earth_temp.txt"
'''
Mean Surface Air Temperature over Ocean or Land Areas (C)
Year,Land_Annual,Lowess(5),Ocean_Annual,Lowess(5)
1880,-0.57,-0.46,-0.05,0.01
1881,-0.36,-0.50,0.01,-0.02
1882,-0.48,-0.54,0.00,-0.06
1883,-0.56,-0.58,-0.06,-0.09
1884,-0.82,-0.61,-0.15,-0.12
1885,-0.75,-0.63,-0.21,-0.15
'''
#df = pd.read_csv(file, delimiter =',', header=[0,1,2]) 


import matplotlib.ticker as ticker

df = pd.read_csv(file, sep=",", header=2)
df.columns = ["x", "y1", "y2", "y3","y4"]
print(df)
```

```python
fig, ax = plt.subplots()

plt.rcParams["figure.figsize"] = (20,12)

ax.plot(df['x'],df['y1'], label='Land annual ', color ='red', lw=1.2 )
ax.plot(df['x'],df['y2'], label='Lowess(5) ?? ', color = 'orange', lw=1.2)
ax.plot(df['x'],df['y3'], label='Ocean annual ', color ='blue', lw  =1.2 )
ax.plot(df['x'],df['y4'], label='Lowess(5) ?? ', color ='cyan', lw= 1.2)
tick_spacing = 30 
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
plt.grid(True, linestyle = '-.' )
plt.xlabel('Year', fontsize = 14)
plt.ylabel('Temperature ($^o$C )', fontsize =14)
plt.title('Mean Temperature ($^o$C ) ', fontsize=16)
plt.legend(loc ='best', fontsize=12)
plt.show()
```
