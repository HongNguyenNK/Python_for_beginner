```python
import matplotlib.pyplot as plt
import pandas as pd

import matplotlib.ticker as ticker
from mpl_toolkits.axes_grid1.inset_locator import zoomed_inset_axes
from mpl_toolkits.axes_grid1.inset_locator import mark_inset 



file = "E:/Documences/py/Dr_Tram/smoothedco2.txt" 


df = pd.read_csv(file, sep="\t", skiprows=1)   
df.fillna(0)
df.columns = ["x", "y1", "y2", "y3", "y4", "y5", "y6", "y7", "y8", "y9", "y10", "y11", "y12", "y13", "y14", "y15", "y16", "y17", "y18", "y19", "y20", "y21", "y22"] 

print(df) 
```

```python
fig, ax = plt.subplots() 

ax.plot(df['x'],df['y1'], label='ALL_50_full', lw=3, color = 'lime', alpha=0.6) 

ax.plot(df['x'],df['y2'], label='LD_050',  lw=1.2, color = 'blue', linestyle ='--')
ax.plot(df['x'],df['y3'], label='DML_050', lw=1.2, color ='blue', linestyle ='-.')
ax.plot(df['x'],df['y4'], label='ALL_050 ',  lw=1.2, color ='blue', linestyle =':')

ax.plot(df['x'],df['y5'], label='LD_075	', lw=1.2, color ='green', linestyle ='--')
ax.plot(df['x'],df['y6'], label='DML_075',  lw=1.2, color ='green', linestyle ='-.')
ax.plot(df['x'],df['y7'], label='ALL_075', lw=1.2, color ='green', linestyle = ':')

ax.plot(df['x'],df['y8'], label='LD_100',  lw=1.2, color ='red', linestyle='--') 
ax.plot(df['x'],df['y9'], label='DML_100', lw=1.2, color = 'red', linestyle ='-.' )
ax.plot(df['x'],df['y10'], label='ALL_100',  lw=1.2, color = 'red', linestyle =':')

ax.plot(df['x'],df['y11'], label='LD_125', lw=1.2, color ='cyan', linestyle ='--')
ax.plot(df['x'],df['y12'], label='DML_125',  lw=1.2, color ='cyan', linestyle ='-.')
ax.plot(df['x'],df['y13'], label='ALL_125', lw=1.2, color ='cyan', linestyle =':')

ax.plot(df['x'],df['y14'], label='LD_150',  lw=1.2, color ='orange', linestyle ='--')
ax.plot(df['x'],df['y15'], label='DML_150', lw=1.2, color ='orange', linestyle = '-.')
ax.plot(df['x'],df['y16'], label='ALL_150',  lw=1.2, color ='orange', linestyle=':') 


ax.plot(df['x'],df['y17'], label='LD_175', lw=1.2, color ='indigo', linestyle ='--')
ax.plot(df['x'],df['y18'], label='DML_175',  lw=1.2, color ='indigo', linestyle ='-.')
ax.plot(df['x'],df['y19'], label='ALL_175', lw=1.2, color ='indigo', linestyle =':')

ax.plot(df['x'],df['y20'], label='LD_200',  lw=1.2, color ='deeppink', linestyle ='--')
ax.plot(df['x'],df['y21'], label='DML_200', lw=1.2, color ='deeppink', linestyle = '-.')
ax.plot(df['x'],df['y22'], label='ALL_200',  lw=1.2, color ='deeppink', linestyle=':') 




tick_spacing =  100 
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
ax.grid(True, linestyle = '-.' ) 



x1 = 1450 
x2 = 1750 

# select y-range for zoomed region
y1 = 275
y2 = 285

# Make the zoom-in plot:
axins = zoomed_inset_axes(ax, 3.2  , loc= 2 ) # zoom = 2

axins.plot(df['x'],df['y1'], label='ALL_50_full', lw= 6, color = 'lime', alpha=0.9  ) 
axins.plot(df['x'],df['y2'], label='LD_050',  lw=1.2, color = 'blue', linestyle ='--')
axins.plot(df['x'],df['y3'], label='DML_050', lw=1.2, color ='blue', linestyle ='-.')
axins.plot(df['x'],df['y4'], label='ALL_050 ',  lw=1.2, color ='blue', linestyle =':')

axins.plot(df['x'],df['y5'], label='LD_075	', lw=1.2, color ='green', linestyle ='--')
axins.plot(df['x'],df['y6'], label='DML_075',  lw=1.2, color ='green', linestyle ='-.')
axins.plot(df['x'],df['y7'], label='ALL_075', lw=1.2, color ='green', linestyle = ':')

axins.plot(df['x'],df['y8'], label='LD_100',  lw=1.2, color ='red', linestyle='--') 
axins.plot(df['x'],df['y9'], label='DML_100', lw=1.2, color = 'red', linestyle ='-.' )
axins.plot(df['x'],df['y10'], label='ALL_100',  lw=1.2, color = 'red', linestyle =':')

axins.plot(df['x'],df['y11'], label='LD_125', lw=1.2, color ='cyan', linestyle ='--')
axins.plot(df['x'],df['y12'], label='DML_125',  lw=1.2, color ='cyan', linestyle ='-.')
axins.plot(df['x'],df['y13'], label='ALL_125', lw=1.2, color ='cyan', linestyle =':')

axins.plot(df['x'],df['y14'], label='LD_150',  lw=1.2, color ='orange', linestyle ='--')
axins.plot(df['x'],df['y15'], label='DML_150', lw=1.2, color ='orange', linestyle = '-.')
axins.plot(df['x'],df['y16'], label='ALL_150',  lw=1.2, color ='orange', linestyle=':') 


axins.plot(df['x'],df['y17'], label='LD_175', lw=1.2, color ='indigo', linestyle ='--')
axins.plot(df['x'],df['y18'], label='DML_175',  lw=1.2, color ='indigo', linestyle ='-.')
axins.plot(df['x'],df['y19'], label='ALL_175', lw=1.2, color ='indigo', linestyle =':')

axins.plot(df['x'],df['y20'], label='LD_200',  lw=1.2, color ='deeppink', linestyle ='--')
axins.plot(df['x'],df['y21'], label='DML_200', lw=1.2, color ='deeppink', linestyle = '-.')
axins.plot(df['x'],df['y22'], label='ALL_200',  lw=1.2, color ='deeppink', linestyle=':') 


axins.set_xlim(x1, x2)
axins.set_ylim(y1, y2)
axins.grid(True, linestyle = '-.')   
plt.xticks(visible=False)
plt.yticks(visible=False)
mark_inset(ax, axins, loc1=3 , loc2=4, fc="none", lw=1.2, ec='k')



ax.set_xlabel('Year') 
ax.set_ylabel('CO$_2$ (ppm)')  
plt.title('CO$_2$ has changed over 1000 years', y=1.1 ) 
ax.legend(bbox_to_anchor=(1.04,1),loc="center left")    
plt.show() 
```
