```python
import matplotlib.pyplot as plt
import pandas as pd


file = "E:/Documences/py/Dr_Tram/Earth_temp_seasonal_resolution.txt"
'''
Temperature Anomaly Seasonal Resolution
Year+Season,Global,Low_Latitudes
1880.29,-0.11,-0.04
1880.54,-0.15,-0.21
1880.79,-0.19,-0.1
1881.04,-0.16,0.01
1881.29,0.06,0.31
1881.54,-0.06,0.12
1881.79,-0.17,-0.03
'''
#df = pd.read_csv(file, delimiter =',', header=[0,1,2]) 


import matplotlib.ticker as ticker

df = pd.read_csv(file, sep=",", header=2)
df.columns = ["x", "y1", "y2"]
print(df)
```

```python
fig, ax = plt.subplots()
plt.rcParams["figure.figsize"] = (12,8)

ax.plot(df['x'],df['y1'], label='Global ', color ='red', lw=1.2 )
ax.plot(df['x'],df['y2'], label='Lowlatitude ', color = 'black', lw=1.2)


import numpy as np

x = np.arange(df['x'].size)
fit = np.polyfit(x, df['y1'], 1)
fit_fn = np.poly1d(fit)

ax.plot(df['x'], fit_fn(x), 'blue', linestyle ='-.', label ='Linear regression of global temp. anom.')



tick_spacing = 30 
ax.xaxis.set_major_locator(ticker.MultipleLocator(tick_spacing))
plt.grid(True, linestyle = '-.' )
plt.xlabel('Year')
plt.ylabel('Temperature anomalies ($^o$C )')
plt.title('Temperature Anomaly Seasonal Resolution  ')
plt.legend()
plt.show()
```
